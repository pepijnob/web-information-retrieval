```python
#This find the most popular artists based on location (query = country) from the lastfm api

def last_search(query):
    url = 'http://ws.audioscrobbler.com/2.0/?method=geo.gettopartists'
    locality = '&country=' + query
    key = '&api_key=52c68179970c2d234ef1b99f5aff8713'
    form = '&format=json'
    final_url = url + locality + key + form
    response = urllib.request.urlopen(final_url).read()
    json_obj = str(response, 'utf-8')
    data = json.loads(json_obj)
    data = json_normalize(data['topartists'], record_path='artist' )
    #json_obj = urllib2.urlopen(final_url)
    #data = json.load(json_obj)
   
    
    print (data['listeners'], data['name'] )
```
